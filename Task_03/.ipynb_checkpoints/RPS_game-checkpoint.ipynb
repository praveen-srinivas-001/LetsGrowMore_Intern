```python
import random
import tkinter as tk
from tkinter import messagebox

def is_win(player, opponent):
    return (player == 'r' and opponent == 's') or (player == 's' and opponent == 'p') or (player == 'p' and opponent == 'r')

def play(user_choice):
    computer_choice = random.choice(['r', 'p', 's'])
    if user_choice == computer_choice:
        result = 0  # tie
    elif is_win(user_choice, computer_choice):
        result = 1  # user wins
    else:
        result = -1  # computer wins
    update_result(result, user_choice, computer_choice)

def update_result(result, user_choice, computer_choice):
    if result == 0:
        result_label.config(text=f"It's a tie! You both chose {user_choice}")
    elif result == 1:
        global player_score
        player_score += 1
        result_label.config(text=f"You won! You chose {user_choice} and the computer chose {computer_choice}")
    else:
        global computer_score
        computer_score += 1
        result_label.config(text=f"You lost! You chose {user_choice} and the computer chose {computer_choice}")
    update_score()

def update_score():
    score_label.config(text=f"Your Score: {player_score}  Computer Score: {computer_score}")

def reset_game():
    global player_score, computer_score
    player_score = 0
    computer_score = 0
    result_label.config(text="Let's play again!")
    update_score()

# Initialize scores
player_score = 0
computer_score = 0
root = tk.Tk()
root.title("Rock Paper Scissors Game")

# Create buttons for Rock, Paper, Scissors
rock_button = tk.Button(root, text="Rock", command=lambda: play('r'))
rock_button.pack(pady=10)

paper_button = tk.Button(root, text="Paper", command=lambda: play('p'))
paper_button.pack(pady=10)

scissors_button = tk.Button(root, text="Scissors", command=lambda: play('s'))
scissors_button.pack(pady=10)

# Label to show results
result_label = tk.Label(root, text="Make your choice!")
result_label.pack(pady=20)

# Label to show scores
score_label = tk.Label(root, text=f"Your Score: {player_score}  Computer Score: {computer_score}")
score_label.pack(pady=10)

# Reset button
reset_button = tk.Button(root, text="Reset Game", command=reset_game)
reset_button.pack(pady=20)

root.mainloop()
```
