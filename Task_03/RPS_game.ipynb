```python
import random
import tkinter as tk
from tkinter import messagebox

def is_win(player, opponent):
    return (player == 'r' and opponent == 's') or (player == 's' and opponent == 'p') or (player == 'p' and opponent == 'r')

def play(user_choice):
    global rounds_played, player_wins, computer_wins
    computer_choice = random.choice(['r', 'p', 's'])
    
    if user_choice == computer_choice:
        result = 0  # tie
    elif is_win(user_choice, computer_choice):
        result = 1  # user wins
        player_wins += 1
    else:
        result = -1  # computer wins
        computer_wins += 1
    
    rounds_played += 1
    update_result(result, user_choice, computer_choice)
    
    if player_wins > best_of_n // 2 or computer_wins > best_of_n // 2:
        announce_winner()

def update_result(result, user_choice, computer_choice):
    if result == 0:
        result_label.config(text=f"It's a tie! You both chose {user_choice.upper()}", fg='orange')
    elif result == 1:
        result_label.config(text=f"You won! You chose {user_choice.upper()} and the computer chose {computer_choice.upper()}", fg='green')
    else:
        result_label.config(text=f"You lost! You chose {user_choice.upper()} and the computer chose {computer_choice.upper()}", fg='red')
    update_score()

def update_score():
    score_label.config(text=f"Your Score: {player_wins}  Computer Score: {computer_wins}")

def reset_game():
    global player_wins, computer_wins, rounds_played, best_of_n
    player_wins = 0
    computer_wins = 0
    rounds_played = 0
    best_of_n = int(best_of_n_entry.get())
    result_label.config(text="Let's play again!", fg='blue')
    update_score()

def announce_winner():
    if player_wins > computer_wins:
        messagebox.showinfo("Game Over", "Congratulations! You have won the game!")
    else:
        messagebox.showinfo("Game Over", "The computer has won the game. Better luck next time!")
    reset_game()

# Initialize scores and round 
player_wins = 0
computer_wins = 0
rounds_played = 0
best_of_n = 5  # default value

# Create window
root = tk.Tk()
root.title("Rock Paper Scissors Game")
root.configure(bg='#f0f0f0')

# Best of N games
best_of_n_label = tk.Label(root, text="Best of N games:", bg='#f0f0f0', font=('Helvetica', 12))
best_of_n_label.pack(pady=5)

best_of_n_entry = tk.Entry(root, width=5, font=('Helvetica', 12))
best_of_n_entry.pack(pady=5)
best_of_n_entry.insert(0, "5")

# Create buttons for Rock, Paper, Scissors
rock_button = tk.Button(root, text="Rock", command=lambda: play('r'), width=15, bg='lightblue', font=('Helvetica', 14))
rock_button.pack(pady=10)

paper_button = tk.Button(root, text="Paper", command=lambda: play('p'), width=15, bg='lightgreen', font=('Helvetica', 14))
paper_button.pack(pady=10)

scissors_button = tk.Button(root, text="Scissors", command=lambda: play('s'), width=15, bg='lightcoral', font=('Helvetica', 14))
scissors_button.pack(pady=10)

# Label to show results
result_label = tk.Label(root, text="Make your choice!", bg='#f0f0f0', font=('Helvetica', 12))
result_label.pack(pady=20)

# Label to show scores
score_label = tk.Label(root, text=f"Your Score: {player_wins}  Computer Score: {computer_wins}", bg='#f0f0f0', font=('Helvetica', 12))
score_label.pack(pady=10)

# Reset button
reset_button = tk.Button(root, text="Reset Game", command=reset_game, width=15, bg='yellow', font=('Helvetica', 14))
reset_button.pack(pady=20)

root.mainloop()
```
